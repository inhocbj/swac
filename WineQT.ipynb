### 상관분석 결과 0.2이상의 값을 가지는 독립변수 선정해서 모델링 진행

```python
X = df.loc[:,['volatile acidity','citric acid','sulphates','alcohol']]
Y = df.loc[:,'quality']
```

### 3. 선형회귀 모델링 사용

```python
lr = linear_model.LinearRegression()
lr.fit(X,Y)
```

```python
print("회귀계수", lr.coef_)
print("회귀상수",lr.intercept_)
```

```python
parmas = np.append(lr.intercept_,lr.coef_)
desighX = np.append(np.ones((len(X),1)),X,axis=1)
predictions = lr.predict(X)
MSE = (sum((Y-predictions)**2))/(len(desighX)-len(desighX[0]))
print(MSE)
```

```python
var_b = MSE*(np.linalg.inv(np.dot(desighX.T,desighX)).diagonal())
se_b = np.sqrt(var_b)
print(se_b)
```

```python
ts_b = parmas/se_b
print(ts_b)
```

```python
from scipy import stats
p_values =[2*(1-stats.t.cdf(np.abs(i),(len(desighX)-len(desighX[0])))) for i in ts_b]
print(p_values)
```

```python
myDF3 = pd.DataFrame()
myDF3["Coefficients"],myDF3["Standard Reeors"],myDF3["t values"],myDF3["p values"] = [parmas,se_b,ts_b,p_values]
print(myDF3)
```

### 종속변수가 등급을 나타내는 변수여서 모형을 변경해서 진행

## 4. LogisticRegression 모델링
- 데이터셋도 학습과 테스트로 나눠서 진행
- 크기 문제로 등급을 5,6 등급만 별도로 추출해서 데이터 사용

```python
import numpy as np
import pandas as pd
import random
from matplotlib import pyplot as plt
from sklearn import linear_model

df = pd.read_csv('./WineQT.csv')

df1 = df[(df['quality'] >= 5) & (df['quality'] <=6)  ]
data = df1.loc[:,['volatile acidity','citric acid','sulphates','alcohol']]
target = df1['quality']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size = 0.3, random_state = 42)
display(X_train, X_test, Y_train, Y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
print(lr.score(X_train, Y_train))
confusion_matrix(lr.predict(X_train),Y_train)
```

```python
print(lr.score(X_test, Y_test))
confusion_matrix(lr.predict(X_test),Y_test)
```
